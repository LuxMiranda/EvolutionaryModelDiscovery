```python
!pip install py4j
import py4j.java_gateway as jg
from py4j.java_gateway import JavaGateway
gw = None    
gw = JavaGateway()# New gateway connection    
def run_model(model_string, steps, metric):
    bridge = gw.entry_point
    #bridge.refresh()
    session = bridge.openModel(model_string)
    setParamsRandom(session,gw,bridge, model_string)
    bridge.command(session,"setup")
    bridge.command(session,"repeat " + str(steps) + " [go]")
    result = bridge.report(session,str(metric))
    bridge.closeModel(session)
    return result

def disconnect():
    gw.shutdown(True)

def setParamsRandom(session,gateway, bridge, model_string):
    paramSpecs = bridge.getParamList(session,model_string).getParamSpecs()

    for paramSpec in paramSpecs:
        if jg.is_instance_of(gateway,paramSpec,"bsearch.space.DoubleDiscreteSpec"):
            paramValue = paramSpec.generateRandomValue(gateway.jvm.org.nlogo.api.MersenneTwisterFast())
        if jg.is_instance_of(gateway,paramSpec,"bsearch.space.DoubleContinuousSpec"):
            paramValue = paramSpec.generateRandomValue(gateway.jvm.org.nlogo.api.MersenneTwisterFast())
        if jg.is_instance_of(gateway,paramSpec,"bsearch.space.CategoricalSpec"):
            paramValue = paramSpec.generateRandomValue(gateway.jvm.org.nlogo.api.MersenneTwisterFast())
            if type(paramValue) != bool:#isinstance(data[i][k], bool)
                paramValue = '"{}"'.format(paramSpec.generateRandomValue(gateway.jvm.org.nlogo.api.MersenneTwisterFast()))
        if jg.is_instance_of(gateway,paramSpec,"bsearch.space.ConstantSpec"):
            paramValue = paramSpec.generateRandomValue(gateway.jvm.org.nlogo.api.MersenneTwisterFast())
        print("set " + str(paramSpec.getParameterName()) + " " + str(paramValue))
        bridge.command(session,"set " + str(paramSpec.getParameterName()) + " " + str(paramValue))
    bridge.command(session,"set " + str(paramSpec.getParameterName()) + " " + str(paramValue))
    bridge.command(session,"set historic-view? true" )
```

```python
def emd_run(): 
    metric = "count households"
    model_string = "C:\\Users\\ch328575\\Documents\\Projects\\EMD\\EMD1.0\\models\\ArtificialAnasazi\\Artificial Anasazi Ver 6.nlogo"

    print(run_model(model_string, 200, metric))
thread_1 = threading.Thread(target=emd_run, args=())
thread_2 = threading.Thread(target=emd_run, args=())
thread_1.start()
thread_2.start()
```
